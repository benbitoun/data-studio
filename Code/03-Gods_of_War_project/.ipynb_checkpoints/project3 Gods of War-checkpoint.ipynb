```python
params = {"figure.facecolor": "#cad9e1",
              "axes.facecolor": "#cad9e1",
              "axes.grid" : True,
              "axes.grid.axis" : "y",
              "grid.color"    : "#ffffff",
              "grid.linewidth": 2,
              "axes.spines.left" : False,
              "axes.spines.right" : False,
              "axes.spines.top" : False,
              "ytick.major.size": 0,     
              "ytick.minor.size": 0,
              "xtick.direction" : "in",
              "xtick.major.size" : 7,
              "xtick.color"      : "#191919",
              "axes.edgecolor"    :"#191919",
              "axes.prop_cycle" : plt.cycler('color',
                                    ['#006767', '#ff7f0e', '#2ca02c', '#d62728',
                                     '#9467bd', '#8c564b', '#e377c2', '#7f7f7f',
                                     '#bcbd22', '#17becf'])}
plt.rcParams.update(params)

ax = s_america2.plot(x='Year', figsize=(30,15))
#Getting rid of y-ax
ax.grid(False, axis='x')


ax.tick_params(left=True)
ax.tick_params(bottom=True)
ax.set_xticks([1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2017])

#Title
ax.set_title("Gods of War Part I", pad=40, fontsize=30)
plt.gcf().text(0.5, 0.9, "U.S Weapon Exports to South American Regimes 1950 to 2017 (in m TIV*) ", ha="center", fontsize=14)

#Save
plt.savefig('south_america_total.pdf', transparent=True, bbox_inches='tight')
plt.savefig('south_america_total.png', transparent=True, bbox_inches='tight')
```

### Playing aroung with an area chart 

```python

params = {"figure.facecolor": "#2A2A2A",
              "axes.facecolor": "#2A2A2A",
              "axes.grid" : True,
              "axes.grid.axis" : "y",
              "grid.color"    : "grey",
              "grid.linewidth": 2,
              "axes.spines.left" : False,
              "axes.spines.right" : False,
              "axes.spines.top" : False,
              "ytick.major.size": 0,     
              "ytick.minor.size": 0,
              "xtick.direction" : "in",
              "xtick.major.size" : 7,
              "xtick.color"      : "#191919",
              "axes.edgecolor"    :"#191919",
              "axes.prop_cycle" : plt.cycler('color',
                                    ['#4490A7', '#43987D', '#AA64EA', '#DC4F7F',
                                     '#CD8342', '#81C885', '#3D9BB9', '#969696',
                                     '#fcae91', '#17becf'])}
plt.rcParams.update(params)

ax = s_america2.plot(kind= 'area', stacked=True, x='Year', figsize=(30,15)) 
#Getting rid of y-ax
ax.grid(False, axis='x')

#ticks
ax.tick_params(left=True)
ax.tick_params(bottom=True)
ax.set_xticks([1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2017])

ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')


#Title
ax.set_title("Gods of War Part I", pad=40, fontsize=30, color='white')
plt.gcf().text(0.5, 0.9, "U.S Weapon Exports to South American Regimes 1950 to 2017 (in m TIV*) ", ha="center", fontsize=14, color='white')

#Legend
ax.legend(ncol=5, loc='upper center')
#Save
plt.savefig('south_america_total_area.pdf', transparent=True, bbox_inches='tight')
plt.savefig('south_america_total_area.png', transparent=True, bbox_inches='tight')
```
