```python
#For Data:
import pandas as pd
import numpy as np
```

```python
#For plotting:
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
import matplotlib as mpl
```

```python
#Read in the file:
df_pollution = pd.read_excel('air_poll_2016.xlsx',skiprows=1, index_col=0, usecols=[1,3,4,5,6,9,10])
df_pollution.head()
```

```python
#How many cities?
len(df_pollution)
```

```python
#Renaming columns:
df_pollution.rename(columns={'City/Town': 'city',
                             'Country': 'country',
                             
                             'Year': 'year', 
                             'Year.1': 'year2'}, inplace=True)
df_pollution.head()
```

```python
#Renaming first didn't work for all columns:
df_pollution.rename(columns={'Annual mean, ug/m3': 'mean_pm10_ug_per_m3', 'Annual mean, ug/m3.1': 'mean_pm2.5_ug_per_m3'}, inplace=True)
df_pollution.head()
```

```python
#New column:
df_pollution['city_country'] = df_pollution['city'] + ', ' + df_pollution['country']
df_pollution
```

```python
#drop column city, country:
#df_pollution.drop(['city, country'],  axis=1, inplace=True)
#df_pollution.head()
```

```python
#Change order:
df_pollution = df_pollution[['country', 'city', 'city_country', 'mean_pm10_ug_per_m3', 'year', 'mean_pm2.5_ug_per_m3', 'year2']]
df_pollution.head()
```

```python
df_pollution.dtypes
```

```python
#Change dtypes
df_pollution['mean_pm10_ug_per_m3'] = df_pollution['mean_pm10_ug_per_m3'].astype(int)
df_pollution['year'] = df_pollution['year'].astype(int)
df_pollution['mean_pm2.5_ug_per_m3'] = df_pollution['mean_pm2.5_ug_per_m3'].astype(int)
df_pollution['year2'] = df_pollution['year2'].astype(int)
```

```python
df_pollution.dtypes
```

```python
#Save as cleaned csv:
df_pollution.to_csv('city_pollution_cleaned.csv', index=False)
```

```python
#Top 10 most pollutet cities in the world (pm10):
def get_color(city_country):
    if city_country == "Onitsha, Nigeria":
        return 'darkred'
    else:
        return 'lightgrey'

bar_colors = [get_color(name) for name in top_pm10_sorted.city_country]

ax = top_pm10_sorted.plot(x='city_country', y='mean_pm10_ug_per_m3', kind='barh', color=bar_colors, legend=False, figsize=(20,20))
ax.set_xticks([0, 20, 100, 200, 300, 400, 500, 600])
ax.set_xticklabels(["", 'WHO limit', 100, 200, 300, 400, 500, 600])
ax.set_xlabel("particulate pollution (pm 10) in microgram per cubic meter")
ax.grid(True, axis='y', color='white', linestyle=':', linewidth=1)
ax.axvline(20, color='darkblue')


#Title
ax.set_title("Top ten cities where the smog is killing you", pad=30)
ax.title.set_fontsize(50)
ax.title.set_fontname("Impact")

#Save as pdf to work on it in illustrator:
plt.savefig("top10_pm10.pdf")
```
