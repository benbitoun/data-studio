```python
#Funktion für Vorstösse:
import re
def get_vorstoss(row):
    if re.findall(r"[\w]* Motion ", row['Traktandierungstitel']):   
        print('Motion')
        return 'Motion'
    
    elif re.findall(r"Motion ", row['Traktandierungstitel']):   
        print('Motion')
        return 'Motion'
    
    elif re.findall(r"[\w]* Postulat ", row['Traktandierungstitel']):   
        print('Postulat')
        return 'Postulat'
    
    elif re.findall(r"Postulat ", row['Traktandierungstitel']):   
        print('Postulat')
        return 'Postulat'
    
    elif re.findall(r"[\w]* Interpellation ", row['Traktandierungstitel']):   
        print('Interpellation')
        return 'Interpellation'
    
    elif re.findall(r"Interpellation ", row['Traktandierungstitel']):   
        print('Interpellation')
        return 'Interpellation'
    
    elif re.findall(r"[\w]* Anfrage ", row['Traktandierungstitel']):   
        print('Kleine Anfrage')
        return 'Kleine Anfrage'
    
    else:
        print('damn')
```

```python
df['vorstoesse'] = df.apply(get_vorstoss,axis=1)
df.head()
```
